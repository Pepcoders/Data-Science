```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.head()
```

```python
print(cancer.DESCR)
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled = scaler.fit_transform(df)
scaled
```

```python
answer = cancer['target']
```

```python
answer
```

```python
features = scaled.T
cov_matrix = np.cov(features)
```

```python
cov_matrix.shape
```

```python
values , vectors = np.linalg.eig(cov_matrix)
values
```

```python
vectors
```

```python
percentage = []
for value in values:
  percentage.append((value/np.sum(values))*100)
print(percentage)
```

```python
data1 = np.dot(scaled,vectors.T[0])
data2 = np.dot(scaled,vectors.T[1])
data3 = np.dot(scaled,vectors.T[2])
```

```python
result = pd.DataFrame(data1,columns=['PC1'])
```

```python
result['PC2']= data2
```

```python
result['PC3'] = data3
```

```python
result.head()
```

```python
result['ans'] = answer
```

```python
result.head()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
sb.scatterplot(result['PC1'],result['PC2'],hue=result['ans'],s=200)
```

```python
new = result.drop(['ans'],axis=1)
```

```python
new.head()
```

```python
from sklearn.linear_model import LogisticRegression
lr =LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(new,answer,random_state=42,test_size=0.33)
```

```python
lr.fit(xtrain,ytrain)
```

```python
lr.score(xtest,ytest)
```

```python
x_train,x_test,y_train,y_test = train_test_split(scaled,answer,random_state =42,test_size = 0.33)
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```
