```python
## Importing Dependencies

import gym
import highway_env
from stable_baselines3 import PPO
```

```python
env = gym.make("roundabout-v0")
```

```python
for _ in range(5):
    done = False
    env.reset()
    while not done:
        env.render()
        action = env.action_space.sample()
        next_state, reward, done, info = env.step(action)
        print(info)
        print(done)
env.close()
```

```python
model = PPO('MlpPolicy', env, verbose=1)
model.learn(total_timesteps=5000)
```

```python
model.save('model/roundabout')
```

```python
del model
```

```python
model = PPO.load('model/roundabout')
```

```python
## Visualizing the model
for _ in range(10):
    done = False
    obs = env.reset()
    while not done:
        env.render()
        action, _states = model.predict(obs)
        next_state, reward, done, info = env.step(action)
        print(info)
        print(done)
env.close()
```
