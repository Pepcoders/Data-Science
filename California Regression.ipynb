```python
x_train_full, x_test, y_train_full,y_test = train_test_split(housing.data,housing.target,random_state=42)
```

```python
x_train,x_val,y_train,y_val = train_test_split(x_train_full,y_train_full,random_state=42)
```

```python
std = StandardScaler()
```

```python
x_test_scaled,x_val_scaled,x_train_scaled = std.fit_transform(x_test) , std.fit_transform(x_val), std.fit_transform(x_train)
```

```python
model = keras.models.Sequential([
                                 keras.layers.Dense(30,activation='relu'),
                                 keras.layers.Dense(1)
])
```

```python
model.compile(loss = 'mean_squared_error',optimizer = 'adam')
```

```python
apple = model.fit(x_train_scaled,y_train,epochs=30,batch_size=16,validation_data=(x_val_scaled,y_val),validation_batch_size=8)
```

```python
mse_test = model.evaluate(x_test_scaled,y_test)
```
