```python
import pandas as pd
```

```python
df=pd.read_csv("drive/MyDrive/Advertising.csv")
```

```python
df.isna().sum()
```

```python
df.plot(kind="scatter",x="TV",y="sales")
df.plot(kind="scatter",x="radio",y="sales")
df.plot(kind="scatter",x="newspaper",y="sales")
```

```python
feature_col=['TV']
X=df[feature_col]
y=df['sales']

from sklearn.linear_model import LinearRegression

lir = LinearRegression()
```

```python
lir.fit(X,y)
```

```python
lir.coef_
```

```python
lir.intercept_
```

```python
X_new =pd.DataFrame({'TV':[df.TV.min(),df.TV.max()]})

X_new
```

```python
preds=lir.predict(X_new)
```

```python
preds
```

```python
import matplotlib.pyplot as plt


df.plot(kind="scatter",x="TV",y="sales")
plt.plot(X_new,preds,c='r')
```
