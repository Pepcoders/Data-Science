```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
cancer.feature_names
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
scaled= scalar.fit_transform(df)
```

```python
ans = cancer['target']
```

```python
ans
```

```python
features = scaled.T
cov_matrix = np.cov(features)
cov_matrix[:5]
```

```python
values, vectors = np.linalg.eig(cov_matrix)
values.shape
```

```python
vectors.shape
```

```python
explained_variances = []
for value in values:
    explained_variances.append((value / np.sum(values)*100))
 
print(np.sum(explained_variances), '\n', explained_variances,'\n')
```

```python
with plt.style.context('dark_background'):
    plt.figure(figsize=(6, 4))

    plt.bar(range(30), explained_variances, alpha=0.5, align='center',
            label='individual explained variance')
    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal components')
    plt.legend(loc='best')
    plt.tight_layout()
```

```python
projected_1 = scaled.dot(vectors.T[0])
projected_2 = scaled.dot(vectors.T[1])
projected_3 = scaled.dot(vectors.T[2])
res = pd.DataFrame(projected_1, columns=['PC1'])
res['PC2'] = projected_2
res['PC3'] = projected_3
y=cancer['target']
res['Y'] = y
res.head()
projected_1
```

```python
import seaborn as sns
plt.figure(figsize=(20, 10))
sns.scatterplot(res['PC1'], res['PC2'], hue=res['Y'], s=200)
```

```python
new = res.drop(['Y'],axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr= LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(new,y,random_state = 42,train_size=.33)
```

```python
lr.fit(xtrain,ytrain)
```

```python
lr.score(xtest,ytest)
```
