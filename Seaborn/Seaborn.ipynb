There are different Types of Plot for different Type of Data.



1.   Numerical Plot
    *   Relplot
    *   Scatter Plot
    * Line Plot
2.   Categorical Plot
    *   CatPlot
    *   Box Plot
    * Swarm Plot
    * Strip Plot

3. Distribution of Data
    * DistPlot
    * KDE plot
    * Joint Plot 

For Linear regression or RelationShip we have
* RegPlot
* LMPlot




```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
type(tips)
```

```python
tips.head()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex',style='time')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex',style='time',size='size')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex',style='time',size='size',kind='line')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',col='time')
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
#Bydefault jitter = true
sns.catplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
## Swarm Plot????

sns.catplot(x='day',y='total_bill',data=tips,kind='swarm',hue='size')
```

```python
sns.catplot(x='smoker',y='tip',data=tips,order=['No','Yes'],)
```

```python
sns.catplot(x='day',y='tip',kind='box',hue='sex',data=tips)
```
